# Permutation and Combinations

The number of ways in which we can arrange a set of elements

# Permutation

We are finding the number of possible arrangements of items
Order matters (arrangements)
Meaning: How many ways you can arrange items in order
Eg = Like insta username
Eg2 = Password setting:
A password 123 is different from 321. (Same digits, but order matters)
Eg3 = Seating arrangement:
You have 3 friends and 3 chairs. The order in which they sit makes a different arrangement

# Combination

We are finding the number of possible selections of items.
Order doesnot matter (Selections)
Meaning: How many ways you can select items from a group
Eg = Like choosing people for a team who you choose matters, not the order
Eg2 = Lottery ticket
picking number {3, 5, 7} is the same as {7, 5, 3}

Eg3 = choosing players for a cricket team:
If you select 2 players from 5, it doesn't matter who's picked first - only who's in the team

```python
from itertools import permutations   # iterator tool


# All permutations of 4 names (default length = 4)
seq = list(permutations(["Alex", "Bhavesh", "Chandrika", "Rohan"]))

seq
```

```python
from itertools import permutations, combinations

letters = ['p', 'y', 't', 'h', 'o', 'n']

# Permutations of length 2 (order matter)
perm = list(permutations(letters, 2))
print("Permutations (Order matters)")
print(perm)
print("Total Permutation:", len(perm))




# Combinations of length 2 (order does not matter)
combi = list(combinations(letters, 2))
print("Combinations (order does not matter):")
print(combi)
print("Total combinations:", len(combi))
```

```python
from scipy import stats

x = [0, 1]
p = 0.7

stats.bernoulli.pmf(x, p)    # Here 0.7 is the number of survived and 0.3 is death 
```

```python
x = [0, 1]
p = 0.56

# CDF = P(X <= x)

stats.bernoulli.cdf(x, p)   # Here 1 is the Total probability
```

```python
# Generate a random space having 100 values in a bernoulli distribution with p = 0.75

p = 0.75               # If i decrease the value of probability to 25 % so i will get more value of 0
rs = stats.bernoulli.rvs(p, size = 100)
print(rs)
```

```python
prob = stats.binom.pmf(10, 30, 0.20)

percentage = prob * 100
print(prob)
print(percentage)
```

```python
prob = stats.binom.cdf(10, 30, 0.2)

percentage = prob * 100
print(prob)
print(percentage)
```

```python
stats.binom.stats(30, 0.2, moments='mvsk')
```

```python
mean, var, skew, kurt = stats.binom.stats(30, 0.2, moments = 'mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```
