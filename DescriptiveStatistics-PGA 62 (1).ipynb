### Skewness p

Skewness is used to measure the asymmetry in our data. It is the measure of asymmetry that occurs when the data deviates from the normal distribution.  

Measure of skewness helps us to know to what degree and in which direction (positive or negative) the frequency distribution has departure from symmetry.

Positive or Negative skewnesscan be detected graphically depending upon whether the right tail or left tail is longer but, we don't get the idea of magnitude.

If mean = 100 , mode = 80 , and std deviation = 20. Find coeff of skewness.

```python
#P
#0earson’s coefficient of skewness is a measure of how asymmetric (skewed) a data distribution is around its mean


# Pearson’s skewness (using Mode)
#skew_mode = (mean - mode) / std

# Pearson’s skewness (using Median)
#skew_median = (3 * (mean - median)) / std

mean = 100
mode = 80
median = 90

std_dev = 20

coeff_skw = (mean - mode)/std_dev
print(coeff_skw)


#The data is positively skewed (right-skewed).
#The tail on the right side of the distribution is longer
```

# If mean = 50 , var = 400 & coeff_skew = -0.4 . Find mode & median.

```python
# Given values
mean = 50
variance = 400
std_dev = math.sqrt(variance)  # Standard deviation
coeff_skew = -0.4

# Formula: Sk = (Mean - Mode) / Std_dev

# Rearranging:
# Mean - Mode = Sk * Std_dev
# Mode = Mean - (Sk * Std_dev)

mode = mean - coeff_skew * std_dev




#empirical relation is an approximate relationship between the mean, median, and mode given by Karl Pearson:
# it is not a strict mathematical law, but an observed relationship

# Empirical relation: Mean - Mode = 3 * (Mean - Median)

# Rearranging:
# Median = Mean - (Mean - Mode) / 3
median = mean - (mean - mode) / 3

# Display results
mode, median

```

### Kurtosis
Kurtosis is also one of the measure of the shape of a frequency curve.Its a Greek word, which means bulginess.

It also helps to denote the heaviness of the distribution towards the tail. While skewness signifies extent of asymmetry, kurtosis measure the degree of peakness or tailedness of a frequency distributi
1. Platykurtic Distribution: This type of distribution has a lower peak and a wider spread than a normal distribution. A platykurtic distribution is characterized by a small number of outliers and can indicate that the data is spread out over a large range. An example of a platykurtic distribution is the uniform distribution.
2. 
Mesokurtic Distribution: A mesokurtic distribution has the same peak and spread as a normal distribution. This type of distribution is common in many real-world scenarios, and the data is usually distributed symmetrically around the mea.
3. L
Leptokurtic Distribution: A leptokurtic distribution has a higher peak and a narrower spread than a normal distribution. A leptokurtic distribution has a higher concentration of data around the mean, and it can indicate that the data has more outliers than a normal distribution.on.

### pp Coefficient of Variation

Coefficient of variation of data defines a parameter to measure the consistency in the data.   
The coefficient of variation is a relative measure of dispersion. It is based on the mean & standard deviation of a frequency distribution.  
It is also known as relative std deviation.  
More the value of C.V. = Less is the consistency.   
Less the value of C.V. = More is the consistency.   


The coefficient of variation is a dimensionless relative measure of dispersion defined as the standard deviation to mean ratio. If two data sets have different units, the coefficient of variation is the best way to compare them.  

Find the coefficient of variation of the following sample set of numbers.

{1, 5, 6, 8, 10, 40, 65, 88}

```python
import pandas as pd

data = pd.Series([1, 5, 6, 8, 10, 40, 65, 88])

mean_data=data.mean()

std_dev = data.std()

CV = (std_dev/mean_data)*100
print(CV)
```

```python
import pandas as pd
# given series of data
sheldon_sub = pd.Series([3,5.1, 2.9,3.7, 4.2, 4.2, 4.3, 5.2, 3.7,4.4,5.2,3.4])
ross_sub = pd.Series([4, 4.6, 3.1, 4.2, 5.4, 3.4, 5, 6.2, 4.2, 3.5, 4.1, 5.5])
joey_sub = pd.Series([4.3, 4, 4.5, 5.1, 6, 5, 4.8, 5.4, 5.1, 6.4, 4.2, 3.4])

# average new subscribers for sheldon
sheldon_avg = sheldon_sub.mean()

# standard deviation for sheldon
sheldon_sd = sheldon_sub.std()

# calculate c.v. of sheldon
sheldon_cv = (sheldon_sd/sheldon_avg)*100

# average new subscribers for ross
ross_avg = ross_sub.mean()

# standard deviation for ross
ross_sd = ross_sub.std()

# calculate c.v. of ross
ross_cv = (ross_sd/ross_avg)*100

# average new subscribers for joey
joey_avg = joey_sub.mean()

# standard deviation for joey
joey_sd = joey_sub.std()

# calculate c.v. of joey
joey_cv = (joey_sd/joey_avg)*100

# print the cv for each blogger
print('C.V. for Sheldon:', sheldon_cv)
print('C.V. for Ross:', ross_cv)
print('C.V. for Joey:', joey_cv)
```

```python
def CV(s):
    mu=s.mean()
    sd=s.std()
    return(sd/mu)*100
```

#### Z-score
A z-score measures exactly how many standard deviations above or below the mean a data point is.

Some important facts about z-scores:

positive z-score says the data point is above average.  
negative z-score says the data point is below average.  
z-score close to 0, says the data point is close to average.  
A data point can be considered unusual if its z-score is above 3 or below −3

```python
tt=[26, 33, 65, 28, 34, 55, 25, 44, 50, 36, 26, 37, 43, 62, 35, 
    38, 45, 32, 28, 34]
def z_score(data):
    data_mean=np.mean(data)
    data_std=np.std(data)
    std_data=[]
    for i in data:
        new_i=(i-data_mean)/data_std
        std_data.append(new_i)
    print("Mean is:", data_mean)
    print("Std is: ",data_std)
    return std_data
```

```python
print(z_score(tt))
```

```python
a=z_score(tt)
print(np.mean(a))
print(np.std(a))
```
