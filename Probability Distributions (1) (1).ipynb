#### Permutations and Combinations
The number of ways in which we can arrange a set of elements.

```python
Permutation

We are finding the number of possible arrangements of items.
Order matters (arrangements).
Meaning: How many ways you can arrange items in order.
EG=like insta username
EG2=Password setting:
A password 123 is different from 321. (Same digits, but order matters).
EG3=Seating arrangement:
You have 3 friends and 3 chairs. The order in which they sit makes a different arrangement.
```

```python
Combination
We are finding the number of possible selections of items.
Order does NOT matter (selections).
Meaning: How many ways you can select items from a group.
EG=like choosing people for a team who you choose matters, not the order.
EG=Lottery ticket:
Picking numbers {3, 5, 7} is the same as {7, 5, 3}.

EG3=Choosing players for a cricket team:
If you select 2 play'ers from 5, it doesn’t matter who’s picked first — only who’s in the team.
```

Scenario-Use

Discrete (countable outcomes)-PMF for exact value, CDF for ≤ value

Continuous (measurable quantities)-PDF for density at a point, CDF for ≤ value

Single trial success/failure-Bernoulli

Multiple independent trials success/failure	Binomial

Bernoulli = n = 1, two outcomes

Binomial = n > 1, count of successes

PMF = probability of exact value (discrete)

CDF = probability ≤ value (cumulative)

PDF = density for continuous

CMF = same as CDF (discrete)

| Distribution | Random Variable X | PMF / PDF      | CDF              | Random Samples      
 
 Bernoulli     0 or 1             `pmf(k, p)`       `cdf(k, p)`       `rvs(p, size)`      
 Binomial      0…n successes      `pmf(k, n, p)`    `cdf(k, n, p)`    `rvs(n, p, size)`   
 Poisson       0… infiniteevents   `pmf(k, lam)`     `cdf(k, lam)`     `rvs(lam, size)`    
 Normal        Continuous real    `pdf(x, mu, sd)`  `cdf(x, mu, sd)`  `rvs(mu, sd, size)` 


| Distribution          | Parameter | Meaning / Description                                    |
| --------------------- | --------- | -------------------------------------------------------- |
| **Bernoulli**         | `p`       | Probability of success (e.g., customer complains)        |
|                       | `k`       | Specific outcome (0 = failure, 1 = success)              |
| **Binomial**          | `n`       | Number of independent trials (e.g., number of customers) |
|                       | `p`       | Probability of success in each trial                     |
|                       | `k`       | Number of successes you are interested in (0 ≤ k ≤ n)    |
| **Poisson**           | `lam`     | Average number of events in a fixed interval (λ)         |
|                       | `k`       | Number of events for which probability is calculated     |
| **Normal**            | `mu`      | Mean (average) of the distribution                       |
|                       | `sd`      | Standard deviation (spread of the data)                  |
|                       | `x`       | Value at which PDF or CDF is calculated                  |
| **All distributions** | `size`    | Number of random samples to generate using `rvs`         |


k - discrete value/count (used in PMF/PMF for Bernoulli, Binomial, Poisson)

p - probability of success (Bernoulli/Binomial)

n - total trials (Binomial)

lam - expected count per interval (Poisson)

mu - mean (Normal)

sd - standard deviation (Normal)

#### Binomial Distribution
Huge Fruit Basket (HFB) is a grocery shop that sells fruits. It is observed that 20% of their customers complain about the fruits purchased by them for many reasons (bad quality, foul smell and less quantity). On Friday, 30 customers purchased fruits from HFB.

Calculate the probability that exactly 10 customers will complain about the purchased products.


```python
?stats.binom.pmf
```

```python
from scipy import stats



prob = stats.binom.pmf(10, 30, 0.20)   # Probability of exactly 10 successes in 30 trials
#with success probability 0.20 each time.

# n = 30 number of trials
# p = 0.20  probability of success in each trial
# k = 10  we want the probability of getting exactly 10 successes

# Convert probability into percentage
percentage = prob * 100

print(prob)        # Actual probability ( 0.0746)
print(percentage)  # In percentage form ( 7.46%)

# Meaning:
# If you perform 30 trials 
# and the probability of success in each trial is 20%,
# then the chance of getting EXACTLY 10 successes is about 7.46%.
```

```python
from scipy import stats

# Problem setup:
# n = 30 number of customers (trials)
# p = 0.2  probability that a customer complains
# k = 10  we want probability of at most 10 complaints

prob = stats.binom.cdf(10, 30, 0.2)  

# .cdf(k, n, p) = cumulative distribution function
# It calculates: P(X <= k)
#  Probability that 10 or fewer customers complain out of 30.

percentage = prob * 100 

print(prob)         # 0.9529  probability value
print(percentage)  #  95.29  percentage form


# There is about a 95.3% chance that at most 10 customers
# will complain out of the 30 on that Friday.
```

Calculate the probability that maximum of 8 customers will complain about the fruits purchased by them.

```python
#P(X<=8)
from scipy import stats  

# Problem: 20% (p=0.2) of customers complain.  
# On Friday, 30 customers purchased fruits.  
# We need P(X <= 8),  probability that at most 8 customers complain.  

# Using Binomial CDF:
# stats.binom.cdf(k, n, p)  gives P(X <= k)
# n = number of trials (30 customers)
# k = maximum number of complaints (8 customers)
# p = probability of success (complaint = 0.2)

prob = stats.binom.cdf(8, 30, 0.2)   # cumulative probability P(X <= 8)
result = prob * 100                   
print(result)  


#There is an 87% chance that at most 8 customers (0 up to 8) will complain out of the 30 customers.
```

#### Poisson Distribution
The number of customer returns in a retail chain per day follows a poisson distribution at a rate of 25 returns per day.

Calculate the probability that the number of returns exceeds 30 in a day.

```python
# formula
Poisson Probability Formula:

P(X = k) = (λ^k * e^(-λ)) / k!

Where:
- P(X = k) = probability of exactly k events happening
- λ (lambda) = average rate of events (mean)
- e = Euler’s number ( 2.718)
- k = number of events
- k! = factorial of k


The Poisson distribution tells us the probability of observing k events in a fixed interval (time, space, etc.)
when events happen randomly but with a known average rate λ.
```

```python
from scipy import stats

# λ (lambda) = average number of returns per day = 25
# We want P(X > 30)probability that returns exceed 30 in a day

# stats.poisson.cdf(30, 25) gives P(X <= 30) for Poisson(λ=25)
# but we want greater than 30 so So, P(X > 30) = 1 - P(X <= 30)
prob = 1 - stats.poisson.cdf(30, 25) #k=30, average rate =25

# Convert probability to percentage
prob * 100
```

The number of road accidents on the day follow Poisson distribution with mean equals to 3. What is the probability that on a day exactly 1 accident will happen?

```python
?stats.poisson.pmf
```

```python
from scipy import stats  

# Poisson distribution is used for events happening independently at a fixed average rate.
# Here, mean (λ) = 3 accidents/day.

# We want: Probability of exactly 1 accident in a day  P(X = 1)

# pmf (Probability Mass Function) gives the probability of exactly 'k' events.
# Syntax: stats.poisson.pmf(k, λ)
# k = 1 (exactly 1 accident), lamda = 3 (mean accidents)

prob = stats.poisson.pmf(1, 3)  # P(X = 1) when mean = 3
print(prob * 100)  # Convert to percentage


```

#### Normal Distribution/ continuous pp
Assume a normal distribution where the mean clotting time of blood is 7.35 seconds, with a standard deviation of 0.35 seconds. What is the probability that blood clotting time will be less than 7 seconds

```python
from scipy import stats

# Mean  of blood clotting time
mu = 7.35  

# Standard deviation  of blood clotting time
sd = 0.35  

# We want the probability that clotting time is less than 7 seconds
# In Normal distribution, P(X < x) = cumulative probability up to x
# stats.norm.cdf(x, mean, std dv) calculates this area under the curve from -∞ to x
probability = stats.norm.cdf(7, mu, sd)

print("Probability that clotting time < 7 seconds:", probability)
```

```python
from scipy import stats
import numpy as np

# Generate random samples from a normal distribution
# stats.norm.rvs(mean, sstd_dev, size=n_samples)
n = stats.norm.rvs(7.35, 0.35, 20)  # 20 random blood clotting times
#rvs = random variates (samples) from a distribution.
#stats.norm.rvs(mean, std, size) → samples from a normal distribution.

print("Random samples:", n)

# Calculate the mean of these 20 samples
print("Sample mean:", np.mean(n))

# Calculate the standard deviation of these 20 samples
print("Sample standard deviation:", np.std(n))
```

A company makes parts for a machine. The lengths of the parts must be within certain limits or they will be rejected.

Assuming this data is normally distributed and 99.7% of the parts were accepted, what are the limits?

```python
mu=3.1
std=0.005

std_3=mu+(3*std)
std_minus3=mu-(3*std)

print("The acceptable range is:",std_minus3, " to ", std_3)
```

```python
import numpy as np
tt=np.array([26, 33, 65, 28, 34, 55, 25, 44, 50, 36, 26, 37, 43, 
             62, 35, 38, 45, 32, 28, 34])
#Normalization
def normalize(data):
    xmin=min(data)
    xmax=max(data)
    norm_data=[]
    for i in data:
        new_i=(i-xmin)/(xmax-xmin)
        norm_data.append(new_i)
    return norm_data
```

```python
print(normalize(tt))
```

```python
new = normalize(tt)
sns.distplot(new)
```

#### Z-score
A z-score measures exactly how many standard deviations above or below the mean a data point is.

Some important facts about z-scores:

positive z-score says the data point is above average.  
negative z-score says the data point is below average.  
z-score close to 0, says the data point is close to average.  
A data point can be considered unusual if its z-score is above 3 or below −3

```python
tt=[26, 33, 65, 28, 34, 55, 25, 44, 50, 36, 26, 37, 43, 62, 35, 
    38, 45, 32, 28, 34]
def z_score(data):
    data_mean=np.mean(data)
    data_std=np.std(data)
    std_data=[]
    for i in data:
        new_i=(i-data_mean)/data_std
        std_data.append(new_i)
    print("Mean is:", data_mean)
    print("Std is: ",data_std)
    return std_data
```

```python
print(z_score(tt))
```

```python
a=z_score(tt)
print(np.mean(a))
print(np.std(a))
```
