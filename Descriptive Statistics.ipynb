# Statistics

Statistics is the art of learning from data. It is concerned with the collection of data, their susequent description, and their analysis, which often leads to the drawing of comclusions.
Statistics helps in decision making

There are two main branches of Statistics
1) Descriptive stats
2) Inferential stats 


**In Population you collect the data from the entire data or entire observation.
Example: All students in a school, all adults in a country, all trees in a forest.

Example:
All registered voters in a particular city. 
All students enrolled in a specific university. 
All cars manufactured by a particular company in a year. 
Every single grain of sand on a beach. 

and **Sample can be known as subset of Population.
Example: 100 students chosen from a school, a survey of 1000 adults, a selection of trees from a forest.

Example:
A random selection of 500 registered voters from the city. 
A group of 100 students chosen from the university for a survey. 
A quality control check of 50 cars randomly selected from the production line. 
A bucket of sand collected from the beach for analysis.




1) Descriptive Stats:- It is used to summarize the data based on Population.
Because we have entire data it is easier for us to calculate Mean, Median, Mode, Average, Min, Max, IQR
Example:- Can calculate Min, max, average marks from entire class using Descriptive stats

2) Inferential Stats:- It is used to make Prediction or Decision using samples.
Hypothesis Testing, Confidence intervals
Example:- From a class of 30 students if i take a sample of only 10 students and do the analysis of how the entire class perform, i can use Inferential Stats

# Skewness

```python
# Pearson's coefficient of skewness is a measure of how asymmetric(skewed) a data distribution is around its mean

# Pearson's skewness (using mode)
# skew_mode = (mean - mode) / std

# Pearson's skewness (using median)
# skew_median = (3 * (mean - median)) / std

mean = 100
mode = 80
median = 90

std_dev = 20

skew_mode = (mean - mode) / std_dev

print(skew_mode)
# The data is positively skewed (right-skewed)
# The tail on the right side of the distribution is longer


skew_median = (3 * (mean - median)) / std_dev

print(skew_median)
```

```python
# If mean = 60, , median = 75 and var = 900. Find coefficient of skewness
# We have var but not std dev can find it out with math library
```

```python
import math

mean = 60
median = 75
var = 900

std = math.sqrt(var)
print("Std deviation =", std)

skew = (3 * (mean - median)) / std

print("Coeff of skew = ", skew)
```

```python
# If mean = 50, var = 400 and coeff_skew = -0.4. Find mode and median

mean = 50
var = 400
std_dev = math.sqrt(var)
coeff_skew = -0.4

mode = mean - (coeff_skew * std_dev)

median = mean - (coeff_skew * std_dev) / 3

print("Standard Deviation:", std_dev)
print("Mode:", mode)
print("Median:", median)
```

# Emperical Rule of Normal Distribution
It is a rule that explains how data is spread in a normal distribution (bell curve)

# Example:-
Exam scores

Mean = 70 marks, Std Dev = 5 marks

68% of students - scored between 65-75

95% - between 60-80

Almost all - between 55-85

# Emperical Rule Ranges
range_68 = (mean - std, mean + std)        68% of data
range_95 = (mean - 2*std, mean + 2*std)    95% of data
range_997 = (mean - 3*std, mean + 3*std)   99.7% of data

```python
data = [162, 158, 149, 167, 162, 154, 172, 160, 164, 159, 175, 163]

mean = np.mean(data)
std = np.std(data)

# Emperical Rule Ranges
range_68 = (mean - std, mean + std)
range_95 = (mean - 2 * std, mean + 2 * std)
range_997 = (mean - 3 * std, mean + 3* std)

print("Mean", mean)
print("Std Dev", std)
print("65% of range", range_68)
print("95% of range", range_95)
print("99.7% of range", range_997)
```

# Normalization:- 
It means Rescaling the data values into a fixed range, usually [0,1]

Normalization (Min-Max scaling) rescales values between 0 and 1

Example of Normalization

Imagine two students:

Student A's score is in marks (0-100) - 85/100
Student B's score is in marks (0-10) - 8.2/10

If we directly compare 85 and 8.2, it doesn't make sense because they are on different scales.

After Normalization to \[0,1]

Student A - 85/100 = 0.85
Student B - 8.2/10 = 0.82

Now both values are on the same scale, so comparison is fair


# Min - Max Normalization Formula
--> X_norm = (X - X_min) / (X_max - X_min)

```python
def normalize(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))

# Example Data: marks of students
marks = np.array([85, 70, 90, 60, 95])

# Applying Normalization
normalized_marks = normalize(marks)

print("Original Marks", marks)
print("Normalized Marks", normalized_marks)
```

# Z-score
A z-score measures exactly how many std dev above or below the mean a data point is.

Some important facts about z-scores:

positive z-score says the data point is above average/mean
negative z-score says the data point is below average/mean
z-score close to 0, says the data point is close to average/mean
A data point can be considered unusual if its z-score is above 3 or below -3

# Z-score standardization formula
# Z = (X - mean) / std   x = data

```python
data = np.array([10, 12, 14, 16, 18])

mean = np.mean(data)
std = np.std(data)


z_scores = (data - mean) / std

print("Original Data:", data)
print("Z-scores", z_scores)
```
