Q2:- There is container with 4 blue triangles, 5 green squares, and 7 red circles. A single object is drawn at random from the container:
1. The object is not a circle
2. The object is a triangle
3. The object is not a square

```python
triangles = 4
squares = 5
circles = 7
total = triangles + squares + circles

prob_not_circle = (triangles + squares) / total

prob_triangle = triangles / total

prob_not_square = (triangles + circles) / total

print("Total:", total)
print("Probability (Not a Circle):", prob_not_circle)
print("Probability (Triangle):", prob_triangle)
print("Probability (Not a Square):", prob_not_square)
```

There are 200 birds in a zoo. 70 birds are male with brown eyes and 100 birds are female with brown eyes. 20 birds are male with blue eyes and 10 birds are female with blue eyes.
Construct a contiguency table. if a bird is select at random, what is the probability that a bird is:
1. A female
2. A male with brown eyes
3. A female given that it has brown eyes
4. A male given that it has blue eyes
5. A creature with blue eyes given that it's a female

```python
male_brown = 70
female_brown = 100
male_blue = 20
female_blue = 10

male_total = male_brown + male_blue       
female_total = female_brown + female_blue  
brown_total = male_brown + female_brown    
blue_total = male_blue + female_blue       
total = male_total + female_total  

print("Total:", total)
print("Male Total:", male_total)
print("Female Total:", female_total)
print("Brown Total:", brown_total)
print("Blue Total:", blue_total)


p_female = (female_total / total) * 100
p_male_and_brown = (male_brown / total) * 100
p_female_given_brown = (female_brown / float(brown_total)) * 100
p_male_given_blue = (male_blue / float(blue_total)) * 100
p_blue_given_female = (female_blue / float(female_total)) * 100


print("1) Prob(Female) =", p_female)
print("2) Prob(Male and Brown eyes) =", p_male_and_brown)
print("3) Prob(Female that has Brown eyes) =", p_female_given_brown)
print("4) Prob(Male that has Blue eyes) =", p_male_given_blue)
print("5) Prob(Blue eyes and it's Female) =", p_blue_given_female)
```
